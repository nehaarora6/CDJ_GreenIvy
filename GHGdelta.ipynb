```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/Users/ailacqua/Cornell/cdj/aashe.csv')
df = pd.DataFrame(data)
```

```python
x = ['Cornell','Princeton','Penn','Yale','Columbia']
ghg_tot = list(df.iloc[:,5])
y = []
for i in range(0,len(ghg_tot)-1,2):
    y.append((ghg_tot[i]-ghg_tot[i+1])/ghg_tot[i+1] * -100)
print(y)
```

```python
plt.figure(figsize=(10,5))
plt.bar(x, y, color='g')
plt.title("AASHE Stars Ivy League Universities GHG Emissions Percent Reduction relative to BY") 
plt.xlabel("Universities") 
plt.ylabel("Percent reduction in GHG emissions relative to BY")
```
