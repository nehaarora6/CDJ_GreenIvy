```python
import pand as as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/Users/ailacqua/Cornell/cdj/aashe.csv')
df = pd.DataFrame(data)
```

```python
x = list(df.iloc[:,0])
ghg_sc = list(df.iloc[:,1])
ghg_os = list(df.iloc[:,2])
ghg_ie = list(df.iloc[:,3])
ghg_te = list(df.iloc[:,4])
ghg_tot = list(df.iloc[:,5])
print(x)
```

```python
X_axis = np.arange(len(x))
print(X_axis - 0.2)
plt.figure(figsize=(14,8))
plt.bar(X_axis - 0.3, ghg_sc, 0.15, label = 'Gross Scope 1 GHG emissions from stationary combustion') 
plt.bar(X_axis - 0.15, ghg_os, 0.15, label = 'Gross Scope 1 GHG emissions from other sources') 
plt.bar(X_axis, ghg_ie, 0.15, label = 'Gross Scope 2 GHG emissions from imported electricity') 
plt.bar(X_axis + 0.15, ghg_te, 0.15, label = 'Gross Scope 2 GHG emissions from imported thermal energy') 
plt.bar(X_axis + 0.3, ghg_tot, 0.15, label = 'Total GHG emissions from all sources') 
plt.xticks(X_axis, x) 
plt.xlabel("Universities") 
plt.ylabel("Emissions (Metric tons CO2)") 
plt.title("AASHE Stars Ivy League Universities GHG Emissions") 
plt.legend()
plt.show() 
```
