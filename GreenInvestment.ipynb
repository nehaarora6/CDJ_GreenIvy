```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
spending = pd.read_csv('greeninvestment.csv')
```

```python
spending = spending[['School', 'PY Sustainable Investments (USD)']]
spending.head()
```

```python
plt.bar(spending['School'], spending['PY Sustainable Investments (USD)'], color='g')
plt.xlabel('University')
plt.ylabel('Investment in hundered millons (USD)')
plt.title("PY Investment in Sustainability Initiatives USD")
plt.xticks(rotation=45)
plt.ylim(100000000, 400000000)
plt.show()
```
