```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('water.csv')
df = pd.DataFrame(data)
df.head()
```

```python
#water use per campus user
plt.figure(figsize=(14,6))
plt.bar(x=df['Parameter'], height=(df['Total water withdrawal']/df['Total campus users']))
plt.xlabel("Universities")
plt.ylabel("Potable water usage (gallons)")
plt.title("Water use per capita at AASHE-STARS universities")
plt.show()
```

```python
#water use per vegetated acre
plt.figure(figsize=(14,6))
plt.bar(x=df['Parameter'], height=(df['Total water withdrawal']/df['Area of vegetated grounds']))
plt.xlabel("Universities")
plt.ylabel("Potable water usage (gallons)")
plt.title("Water use per vegetated acre at AASHE-STARS universities")
plt.show()
```

```python
#water use per building square foot
plt.figure(figsize=(14,6))
plt.bar(x=df['Parameter'], height=(df['Total water withdrawal']/df['Gross floor area of building space']))
plt.xlabel("Universities")
plt.ylabel("Potable water usage (gallons)")
plt.title("Water use per building square foot at AASHE-STARS universities")
plt.show()
```
