```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/Users/ailacqua/Cornell/cdj/aashe.csv')
df = pd.DataFrame(data)
```

```python
x = list(df.iloc[:,0])
ghg_tot = list(df.iloc[:,5])
pop_tot = list(df.iloc[:,21])
y = []
for i in range(0,len(ghg_tot)):
    y.append(ghg_tot[i]/pop_tot[i])
print(y)
```

```python
plt.figure(figsize=(14,5))
plt.bar(x, y, color='g')
plt.title("AASHE Stars Ivy League Universities GHG Emissions Per Capita") 
plt.xlabel("Universities")
plt.ylabel("Emissions Per Capita (Metric tons CO2 per weighted campus user)")
```
