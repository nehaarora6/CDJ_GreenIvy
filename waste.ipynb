```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
schools = ["Cornell","Princeton","Penn","Yale","Columbia"]
data = pd.read_csv('waste.csv')
df = pd.DataFrame(data)
df
```

```python
#plotting fraction of waste diverted
plt.figure(figsize=(14,6))
plt.bar(x=df['Parameter'], height=(df['Waste diverted']/df['Total waste generated']))
plt.xlabel("Universities")
plt.ylabel("Fraction of total waste diverted")
plt.title("Fraction of waste diverted from landfills at AASHE-STARS uninversities")
plt.show()
```

```python
#plotting fraction of waste diverted
percent_reduction = []

for i in range(0,len(df['Parameter']),2):
    percent_reduction.append(-100*(df['Total waste generated'][i+1]-df['Total waste generated'][i])/df['Total waste generated'][i])
    
plt.bar(x=schools, height=percent_reduction)
plt.xlabel("Universities")
plt.ylabel("Percent reduction in total waste relative to BY")
plt.title("Percent reduction in total waste generated at AASHE-STARS uninversities")
plt.show()
```
